```python
import requests
url = "https://www.lagou.com/jobs/list_python/p-city_252?px=default&gx=&xl=%E6%9C%AC%E7%A7%91&isSchoolJob=1#filterBox"

headers = {
    "user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36",
    "referer":"https://www.lagou.com/jobs/list_python/p-city_252?px=default&gx=&isSchoolJob=1"
}


session = requests.Session()
response = session.get("https://www.lagou.com/jobs/list_python/p-city_252?px=default&gx=&isSchoolJob=1#filterBox")
print(response.cookies)
response = session.post(url,cookies=response.cookies,headers = headers)
print(response.text)
```
